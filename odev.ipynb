Bir arabanın sahip olduğu silindir sayısına ilişkin mpg'yi lineer bir regresyon modeli kullanarak tahmin etmeye çalışırsak ne olur?

Bu, herhangi bir programlama aracında veya hatta bugünlerde Excel'de kolayca yapılabilir. Katsayıyı ve kesmeyi bulup çiziyoruz. Buradaki güzel şey, genel modelin nasıl karar verdiğini (küresel düzeyde) basitçe görebilmemizdir. Aynı zamanda, bireysel tahminler küresel düzeydeki tahminlerle tutarlıdır.

````python
import requests
import pandas as pd
from io import StringIO
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
## Veri kümesinin URL'si
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data' r = requests.get(url)
## İçindeki dosyayı okuyun ve df olarak saklayın
file = r.text.replace("\t"," ") ## list_labels elle yazılmış: list_labels = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin','car name']
## ['mpg', 'silindirler', 'yer değiştirme', 'at gücü', 'ağırlık', 'hızlanma', 'yıl modeli', 'Menşei','araba adı']
df = pd.read_csv(StringIO(file),sep="\s+",header = None,names=list_labels)
##modelleme
regr = linear_model.LinearRegression()
train_x = np.asanyarray(df[['cylinders']])
train_y = np.asanyarray(df[['mpg']])
regr.fit (train_x, train_y)
## Model çizimi
plt.scatter(df.cylinders, df.mpg,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Cylinders")
plt.ylabel("MPG")
````

**Doğrusal bir regresyon modelini açıklama:**

Karmaşık modelleri açıklamak için Shapley değerlerini kullanmadan önce, bunların basit modeller için nasıl çalıştığını anlamak faydalı olacaktır. En basit model türlerinden biri standart lineer regresyondur ve bu nedenle aşağıda California konut veri setinde bir lineer regresyon modeli eğitiyoruz. Bu veri seti, 1990'da Kaliforniya'da 20.640 blok evden oluşuyor; burada amacımız, medyan ev fiyatının doğal günlüğünü 8 farklı özellikten tahmin etmektir:

1-MedInc  - blok grubundaki medyan gelir

2-HouseAge - blok grubundaki medyan ev yaşı

3-AveRooms - hane başına düşen ortalama oda sayısı

4-AveBedrms - hane başına ortalama yatak odası sayısı

5-Population - blok grup popülasyonu

6-AveOccup - hane halkının ortalama sayısı

7-Latitude  - blok grubu enlemi

8-Longitude  - blok grubu boylamı

````python
import pandas as pd
import shap
import sklearn
# klasik bir konut fiyatı veri seti
X,y = shap.datasets.california(n_points=1000)

X100 = shap.utils.sample(X, 100) # 100 arka plan dağıtımı olarak kullanım örneği

# basit bir doğrusal model
model = sklearn.linear_model.LinearRegression()
model.fit(X, y)
````

**Model katsayılarının incelenmesi**

Doğrusal bir modeli anlamanın en yaygın yolu, her özellik için öğrenilen katsayıları incelemektir. Bu katsayılar, girdi özelliklerinin her birini değiştirdiğimizde model çıktısının ne kadar değiştiğini bize söyler:

````python
print("Model coefficients:\n")
for i in range(X.shape[1]):
    print(X.columns[i], "=", model.coef_[i].round(5))
````

Katsayılar, bir girdi özelliğinin değerini değiştirdiğimizde ne olacağını bize anlatmak için harika olsa da, kendi başlarına bir özelliğin genel önemini ölçmek için harika bir yol değildir. Bunun nedeni, her katsayının değerinin girdi özelliklerinin ölçeğine bağlı olmasıdır. Örneğin, bir evin yaşını yıl yerine dakika olarak ölçersek, HouseAge özelliğinin katsayıları 0,0115 / (365∗24∗60) = 2,18e-8 olur. Açıktır ki, bir evin inşa edilmesinden bu yana geçen yıl sayısı, dakika sayısından daha önemli değildir, ancak katsayı değeri çok daha büyüktür. Bu, bir katsayının büyüklüğünün, doğrusal bir modelde bir özelliğin öneminin mutlaka iyi bir ölçüsü olmadığı anlamına gelir.

**Kısmi bağımlılık grafikleri kullanılarak daha eksiksiz bir resim:**

Bir özelliğin bir modeldeki önemini anlamak için, hem bu özelliği değiştirmenin modelin çıktısını nasıl etkilediğini hem de bu özelliğin değerlerinin dağılımını anlamak gerekir. Bunu doğrusal bir model için görselleştirmek için klasik bir kısmi bağımlılık grafiği oluşturabilir ve özellik değerlerinin dağılımını x ekseni üzerinde bir histogram olarak gösterebiliriz:

````python
shap.partial_dependence_plot(
    "MedInc", model.predict, X100, ice=False,
    model_expected_value=True, feature_expected_value=True
)
````

Konut veri setinin satış fiyatını tahmin etmek için temel bir model (RandomForestRegressor) oluşturdum.
Bu, Permütasyon Önemini göstermek için iyi bir veri kümesi örneğidir çünkü bu veri kümesinin birçok özelliği vardır.
Böylece değerleri tahmin ederken hangi özelliklerin etki edip hangilerinin etmediğini görebiliriz.

````python
#BAŞLANGIÇ BİR MODEL OLUŞTURDUM VE VERİ KÜMESİNDE BAZI VERİLERDE MANİPÜLASYON YAPTIM
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
file_path = "../input/train.csv"
housing_data = pd.read_csv(file_path)
y = housing_data.SalePrice
feature_names = [i for i in housing_data.columns if housing_data[i].dtype in [np.int64]]
x = housing_data[feature_names]
x = x.drop(["SalePrice","Id"],axis=1)
train_x, val_x, train_y, val_y = train_test_split(x, y, random_state=1)
model = RandomForestRegressor(random_state=1)
model.fit(train_x, train_y)
````

Şimdi Permütasyon önemini hesaplamak için 'eli5' kütüphanesini kullanıyoruz.

````python
#eli5 kitaplığının NASIL KULLANILACAĞINI GÖSTERMEK
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(model, random_state=1).fit(val_x, val_y)
eli5.show_weights(perm, feature_names = val_x.columns.tolist())
````

# **LIME For Time**

Kireç, açıklanabilirlik alanında popülerlik kazanan ilk tekniklerden biriydi. Kireç Yerel yorumlanabilir model agnostik açıklamalar anlamına gelir. Bu yazıda kirecin arkasındaki metodoloji ele alınmaktadır. Şu anda Lime, tablo verileri, resimler ve metin sınıflandırıcılar için tahminlerin açıklanmasına yardımcı olmaktadır.

Lime temel olarak, makine öğrenimi modelinin tahminlerini yerel olarak taklit etmek üzere eğitilmiş yerel vekil modeller yaratarak modelin davranışının yerel doğrusal bir tahminini vermeye çalışır. Küresel önem, tüm veri setinde ortalama bir etki gösterirken, yerel düzeydeki bir gözlem, her bir değişken tarafından farklı şekillerde etkilenebilir. Bu vekil model, GLM'lerden yerel önemin nasıl değişebileceğini anlamaya çalışan karar ağaçlarına kadar her şey olabilir. Christoph Molnar tarafından yorumlanabilir-ml-kitapta belirtilen genel çerçeve şöyledir:

Kara kutu tahmini hakkında bir açıklama almak istediğiniz ilgi alanınızı seçin
Veri kümenizi bozun ve bu yeni noktalar için kara kutu tahminlerini alın
Yeni örnekleri ilgilenilen örneğe olan yakınlıklarına göre ağırlıklandırın
Varyasyonlarla veri kümesinde ağırlıklı, yorumlanabilir bir model eğitin
Tahmini yerel modeli yorumlayarak açıklayın
Ancak Lime'ın belirli sınırlamaları vardır ve dikkatli kullanılmalıdır (daha fazla ayrıntı için bu makaleye bakın). Bu makalede, kirecin derinliklerine dalmak isteyenler için ilginç deneyler ve daha fazla referans verilmektedir.

````python

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.metrics import accuracy_score as acc
import numpy as np
from lime import explanation
from lime import lime_base
from third_party.lime_for_time.lime_timeseries import LimeTimeSeriesExplanation
import math
````

````python
coffee_train = pd.read_csv('third_party/lime_for_time/coffee_train.csv', sep=',', header=None).astype(float)
coffee_train_y = coffee_train.loc[:, 0]
coffee_train_x = coffee_train.loc[:, 1:]

coffee_test = pd.read_csv('third_party/lime_for_time/coffee_test.csv', sep=',', header=None).astype(float)
coffee_test_y = coffee_test.loc[:, 0]
coffee_test_x = coffee_test.loc[:, 1:]
````

````python

for idx in range(0, len(coffee_train)):
    if coffee_train_y[idx] == 0:
        plt.plot(coffee_train_x.loc[idx, :], 'b')
    else:
        plt.plot(coffee_train_x.loc[idx, :], 'r')
plt.title('Coffee Dataset')
plt.show()
````

image.png

````python
knn = KNN()
knn.fit(coffee_train_x, coffee_train_y)
print('Accuracy KNN for coffee dataset: %f' % (acc(coffee_test_y, knn.predict(coffee_test_x))))
````

Accuracy KNN for coffee dataset: 0.964286


````python
idx = 5
num_features = 10
num_slices = 24
series = coffee_test_x.iloc[idx, :]
print(coffee_test_y.iloc[idx])

explainer = LimeTimeSeriesExplanation(class_names=['0', '1'], feature_selection='auto')
exp = explainer.explain_instance(series, knn.predict_proba, num_features=num_features, num_slices=num_slices, 
                                 replacement_method='total_mean', training_set=coffee_train_x)
exp.as_list()
````
